The following book shows how you can interact with **Dog CEO Zine**: A business and lifestyle magazine for the modern dog. 

Take your time, relax, and discover the different types and breeds of dogs that appear as you include the names of your partner, family, friends, pets, and more.

**Your perfect dog breed is just a woof away**

```python
import urllib.request, urllib.parse, urllib.error
import json

#For Search and Displya image
from PIL import Image
from io import BytesIO
from IPython.display import display

# URL for searching your dog
service_url = 'https://dog.ceo/api/breeds/image/random'

# Start message
print("You have to put a real name, don't put numbers. If you do so, it will be at your own risk.")

while True:
    try:
        # Ask for the user's name
        name = input("What is your name? ")

        # Check if the name contains numbers
        if any(char.isdigit() for char in name):
            print("wow wow wow wow wow, that is not a name... grrrr!!!")
            break

        # Searching from Dog CEO API using JSON Data
        with urllib.request.urlopen(service_url) as response:
            data = json.loads(response.read().decode())

        # Get the image URL from the response
        image_url = data['message']

        # Searching the dog's image
        with urllib.request.urlopen(image_url) as image_response:
            img = Image.open(BytesIO(image_response.read()))

        # Print the message and show the dog's image
        print(f"Hello, {name}! This is today's dog of the day for you:")
        display(img)

    except urllib.error.URLError as e:
        print(f"Failed to retrieve data: {e.reason}")
    else:
        print("Data retrieved successfully.")
```
